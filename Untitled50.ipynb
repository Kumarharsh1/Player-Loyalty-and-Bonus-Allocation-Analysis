```python
import pandas as pd

# Load the sheet "Withdrawal Data" while skipping the first 2 rows
df_withdrawal = pd.read_excel(
    r"C:\Users\Asus\OneDrive\Desktop\Analytics .xlsx", 
    sheet_name="Withdrawal Data", 
    skiprows=2
)

# Inspect the first few rows to confirm header placement
df_withdrawal.head()
```

```python
# Set the first row as the header
df_withdrawal.columns = df_withdrawal.iloc[0]
df_withdrawal = df_withdrawal[1:].reset_index(drop=True)

# Clean column names (remove spaces, etc.)
df_withdrawal.columns = df_withdrawal.columns.str.strip()

# Preview cleaned DataFrame
df_withdrawal.head()
```

```python
# Convert 'User Id' and 'Amount' to numeric, handling potential errors
df_withdrawal['User Id'] = pd.to_numeric(df_withdrawal['User Id'], errors='coerce')
df_withdrawal['Amount'] = pd.to_numeric(df_withdrawal['Amount'], errors='coerce')

# Convert 'Datetime' to datetime objects, handling potential errors
df_withdrawal['Datetime'] = pd.to_datetime(df_withdrawal['Datetime'], errors='coerce')

# Preview after conversions
df_withdrawal.head()
```

```python
Data Preprocessing and Feature Engineering
```

```python
print("Cleaning NaN values from dataframes...")

# df_gameplay
initial_rows_gameplay = len(df_gameplay)
df_gameplay.dropna(subset=['User ID', 'Games Played', 'Datetime'], inplace=True)
removed_rows_gameplay = initial_rows_gameplay - len(df_gameplay)
print(f"Removed {removed_rows_gameplay} rows from df_gameplay due to NaN values in critical columns.")

# df_deposit
initial_rows_deposit = len(df_deposit)
df_deposit.dropna(subset=['User Id', 'Amount', 'Datetime'], inplace=True)
removed_rows_deposit = initial_rows_deposit - len(df_deposit)
print(f"Removed {removed_rows_deposit} rows from df_deposit due to NaN values in critical columns.")

# df_withdrawal
initial_rows_withdrawal = len(df_withdrawal)
df_withdrawal.dropna(subset=['User Id', 'Amount', 'Datetime'], inplace=True)
removed_rows_withdrawal = initial_rows_withdrawal - len(df_withdrawal)
print(f"Removed {removed_rows_withdrawal} rows from df_withdrawal due to NaN values in critical columns.")

print("NaN cleaning complete.")
```

```python
print("Creating 'Date' and 'Time Slot' columns...")

# For df_gameplay
df_gameplay['Date'] = df_gameplay['Datetime'].dt.date
df_gameplay['Time Slot'] = df_gameplay['Datetime'].apply(lambda x: 'S1' if x.hour < 12 else 'S2')

# For df_deposit
df_deposit['Date'] = df_deposit['Datetime'].dt.date
df_deposit['Time Slot'] = df_deposit['Datetime'].apply(lambda x: 'S1' if x.hour < 12 else 'S2')

# For df_withdrawal
df_withdrawal['Date'] = df_withdrawal['Datetime'].dt.date
df_withdrawal['Time Slot'] = df_withdrawal['Datetime'].apply(lambda x: 'S1' if x.hour < 12 else 'S2')

print("Date and Time Slot columns created.")

print("\nFirst 5 rows of df_gameplay with new columns:")
print(df_gameplay.head())
print("\nFirst 5 rows of df_deposit with new columns:")
print(df_deposit.head())
print("\nFirst 5 rows of df_withdrawal with new columns:")
print(df_withdrawal.head())
```

```python
print("Aggregating df_deposit data...")

# Aggregate df_deposit by 'User Id' and 'Date'
df_daily_deposits = df_deposit.groupby(['User Id', 'Date']).agg(
    Daily_Deposit_Amount=('Amount', 'sum'),
    Daily_Deposit_Count=('Amount', 'count')
).reset_index()

# Aggregate df_deposit by 'User Id', 'Date', and 'Time Slot'
df_slot_deposits = df_deposit.groupby(['User Id', 'Date', 'Time Slot']).agg(
    Slot_Deposit_Amount=('Amount', 'sum'),
    Slot_Deposit_Count=('Amount', 'count')
).reset_index()

print("df_deposit aggregation complete.")

print("\nFirst 5 rows of df_daily_deposits:")
print(df_daily_deposits.head())

print("\nFirst 5 rows of df_slot_deposits:")
print(df_slot_deposits.head())
```

```python
print("Aggregating df_withdrawal data...")

# Aggregate df_withdrawal by 'User Id' and 'Date'
df_daily_withdrawals = df_withdrawal.groupby(['User Id', 'Date']).agg(
    Daily_Withdrawal_Amount=('Amount', 'sum'),
    Daily_Withdrawal_Count=('Amount', 'count')
).reset_index()

# Aggregate df_withdrawal by 'User Id', 'Date', and 'Time Slot'
df_slot_withdrawals = df_withdrawal.groupby(['User Id', 'Date', 'Time Slot']).agg(
    Slot_Withdrawal_Amount=('Amount', 'sum'),
    Slot_Withdrawal_Count=('Amount', 'count')
).reset_index()

print("df_withdrawal aggregation complete.")

print("\nFirst 5 rows of df_daily_withdrawals:")
print(df_daily_withdrawals.head())

print("\nFirst 5 rows of df_slot_withdrawals:")
print(df_slot_withdrawals.head())
```

```python
print("Aggregating df_gameplay data...")

# Aggregate df_gameplay by 'User ID' and 'Date'
df_daily_gameplay = df_gameplay.groupby(['User ID', 'Date']).agg(
    Daily_Games_Played=('Games Played', 'sum')
).reset_index()

# Aggregate df_gameplay by 'User ID', 'Date', and 'Time Slot'
df_slot_gameplay = df_gameplay.groupby(['User ID', 'Date', 'Time Slot']).agg(
    Slot_Games_Played=('Games Played', 'sum')
).reset_index()

print("df_gameplay aggregation complete.")

print("\nFirst 5 rows of df_daily_gameplay:")
print(df_daily_gameplay.head())

print("\nFirst 5 rows of df_slot_gameplay:")
print(df_slot_gameplay.head())
```

```python
print("Merging daily aggregated data...")

# Ensure 'User ID' column in df_daily_gameplay is consistent with 'User Id' in other dataframes
df_daily_gameplay = df_daily_gameplay.rename(columns={'User ID': 'User Id'})

# Merge df_daily_deposits and df_daily_withdrawals
df_daily_summary = pd.merge(df_daily_deposits, df_daily_withdrawals, on=['User Id', 'Date'], how='outer')

# Merge with df_daily_gameplay
df_daily_summary = pd.merge(df_daily_summary, df_daily_gameplay, on=['User Id', 'Date'], how='outer')

# Fill NaN values with 0 for activity columns
df_daily_summary = df_daily_summary.fillna(0)

print("Daily aggregated data merged into df_daily_summary.")

print("\nFirst 5 rows of df_daily_summary:")
print(df_daily_summary.head())
```

```python
print("Merging slot aggregated data...")

# Ensure 'User ID' column in df_slot_gameplay is consistent with 'User Id' in other dataframes
df_slot_gameplay = df_slot_gameplay.rename(columns={'User ID': 'User Id'})

# Merge df_slot_deposits and df_slot_withdrawals
df_slot_summary = pd.merge(df_slot_deposits, df_slot_withdrawals, on=['User Id', 'Date', 'Time Slot'], how='outer')

# Merge with df_slot_gameplay
df_slot_summary = pd.merge(df_slot_summary, df_slot_gameplay, on=['User Id', 'Date', 'Time Slot'], how='outer')

# Fill NaN values with 0 for activity columns
df_slot_summary = df_slot_summary.fillna(0)

print("Slot aggregated data merged into df_slot_summary.")

print("\nFirst 5 rows of df_slot_summary:")
print(df_slot_summary.head())
```

```python
print("Creating df_monthly_summary for October...")

# Filter df_daily_summary for October (month 10)
df_october_daily = df_daily_summary[df_daily_summary['Date'].apply(lambda x: x.month == 10)]

# Group by 'User Id' and sum the relevant columns for monthly summary
df_monthly_summary = df_october_daily.groupby('User Id').agg(
    Monthly_Deposit_Amount=('Daily_Deposit_Amount', 'sum'),
    Monthly_Withdrawal_Amount=('Daily_Withdrawal_Amount', 'sum'),
    Monthly_Games_Played=('Daily_Games_Played', 'sum'),
    Monthly_Deposit_Count=('Daily_Deposit_Count', 'sum'),
    Monthly_Withdrawal_Count=('Daily_Withdrawal_Count', 'sum')
).reset_index()

print("df_monthly_summary created.")

print("\nFirst 5 rows of df_monthly_summary:")
print(df_monthly_summary.head())
```

```python
print("\nVerifying df_daily_summary:")
print(df_daily_summary.head())
df_daily_summary.info()
print(df_daily_summary.describe())

print("\nVerifying df_slot_summary:")
print(df_slot_summary.head())
df_slot_summary.info()
print(df_slot_summary.describe())

print("\nVerifying df_monthly_summary:")
print(df_monthly_summary.head())
df_monthly_summary.info()
print(df_monthly_summary.describe())
```

```python
Loyalty Point Calculation
```

```python
import numpy as np

print("Calculating loyalty points...")

# 1. Calculate Net_Deposit_Count for df_slot_summary
df_slot_summary['Net_Deposit_Count'] = np.maximum(df_slot_summary['Slot_Deposit_Count'] - df_slot_summary['Slot_Withdrawal_Count'], 0)

# 2. Calculate Slot_Loyalty_Points for df_slot_summary
df_slot_summary['Slot_Loyalty_Points'] = (
    (0.01 * df_slot_summary['Slot_Deposit_Amount']) +
    (0.005 * df_slot_summary['Slot_Withdrawal_Amount']) +
    (0.001 * df_slot_summary['Net_Deposit_Count']) +
    (0.2 * df_slot_summary['Slot_Games_Played'])
)

# 3. Calculate Net_Deposit_Count for df_monthly_summary
df_monthly_summary['Net_Deposit_Count'] = np.maximum(df_monthly_summary['Monthly_Deposit_Count'] - df_monthly_summary['Monthly_Withdrawal_Count'], 0)

# 4. Calculate Monthly_Loyalty_Points for df_monthly_summary
df_monthly_summary['Monthly_Loyalty_Points'] = (
    (0.01 * df_monthly_summary['Monthly_Deposit_Amount']) +
    (0.005 * df_monthly_summary['Monthly_Withdrawal_Amount']) +
    (0.001 * df_monthly_summary['Net_Deposit_Count']) +
    (0.2 * df_monthly_summary['Monthly_Games_Played'])
)

print("Loyalty point calculations complete.")

# 5. Print the first 5 rows of both df_slot_summary and df_monthly_summary
print("\nFirst 5 rows of df_slot_summary with loyalty points:")
print(df_slot_summary.head())

print("\nFirst 5 rows of df_monthly_summary with loyalty points:")
print(df_monthly_summary.head())
```

```python
print("Calculating loyalty points for specific time slots...")

# Define the specific slots requested
specific_slots = [
    (pd.to_datetime('2022-10-02').date(), 'S1'),
    (pd.to_datetime('2022-10-16').date(), 'S2'),
    (pd.to_datetime('2022-10-18').date(), 'S1'),
    (pd.to_datetime('2022-10-26').date(), 'S2')
]

# Filter df_slot_summary for the specific slots
df_specific_slots_loyalty = df_slot_summary[
    df_slot_summary.apply(
        lambda row: (row['Date'], row['Time Slot']) in specific_slots, axis=1
    )
]

# Display loyalty points for these specific slots
print("\nLoyalty points for specific time slots:")
print(df_specific_slots_loyalty[['User Id', 'Date', 'Time Slot', 'Slot_Loyalty_Points']].sort_values(by=['Date', 'Time Slot', 'User Id']).head(20)) # Display first 20 as there might be many users

# Also, calculate the overall monthly loyalty points for October
print("\nOverall monthly loyalty points for October:")
print(df_monthly_summary[['User Id', 'Monthly_Loyalty_Points']].sort_values(by='Monthly_Loyalty_Points', ascending=False).head())
```

```python
print("Ranking players by Monthly Loyalty Points (with Monthly Games Played as tie-breaker)...")

# Rank players by Monthly_Loyalty_Points, with Monthly_Games_Played as a tie-breaker
df_monthly_ranked = df_monthly_summary.sort_values(
    by=['Monthly_Loyalty_Points', 'Monthly_Games_Played'],
    ascending=[False, False]
).reset_index(drop=True)

print("\nTop 10 players by Monthly Loyalty Points (October, with Games Played tie-breaker):")
print(df_monthly_ranked[['User Id', 'Monthly_Loyalty_Points', 'Monthly_Games_Played']].head(10))
```

```python
print("Calculating average monthly metrics for October...")

# Calculate overall average monthly deposit amount for October
overall_avg_monthly_deposit = df_monthly_summary['Monthly_Deposit_Amount'].mean()
print(f"\nOverall Average Monthly Deposit Amount (October): Rs {overall_avg_monthly_deposit:.2f}")

# Calculate average monthly deposit amount per user for October (already available in df_monthly_summary for each user, so we just need the average of these)
avg_monthly_deposit_per_user = df_monthly_summary['Monthly_Deposit_Amount'].mean()
print(f"Average Monthly Deposit Amount Per User (October): Rs {avg_monthly_deposit_per_user:.2f}")

# Calculate average games played per user for October
avg_monthly_games_played_per_user = df_monthly_summary['Monthly_Games_Played'].mean()
print(f"Average Monthly Games Played Per User (October): {avg_monthly_games_played_per_user:.2f}")
```

```python
Bonus Allocation Strategy (Part B)
```

```python
print("Implementing proportional bonus allocation strategy...")

# 1. Select the top 50 players from df_monthly_ranked
df_top_50_players = df_monthly_ranked.head(50).copy()

# Define the total bonus pool
total_bonus_pool = 50000

# 2. Calculate the total loyalty points for these top 50 players
total_loyalty_points_top_50 = df_top_50_players['Monthly_Loyalty_Points'].sum()

# 3. Calculate the proportion of total loyalty points for each of the top 50 players
df_top_50_players['Loyalty_Point_Proportion'] = df_top_50_players['Monthly_Loyalty_Points'] / total_loyalty_points_top_50

# 4. Allocate the bonus amount proportionally
df_top_50_players['Bonus_Amount'] = df_top_50_players['Loyalty_Point_Proportion'] * total_bonus_pool

# 5. Print the 'User Id', 'Monthly_Loyalty_Points', and 'Bonus_Amount' for the top 50 players
print("\nBonus allocation for the top 50 players:")
print(df_top_50_players[['User Id', 'Monthly_Loyalty_Points', 'Bonus_Amount']])
```
